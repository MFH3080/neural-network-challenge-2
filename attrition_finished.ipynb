## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
```

```python
# Create a list of at least 10 column names to use as X data
columns = ['YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsAtCompany', 'JobSatisfaction', 'EnvironmentSatisfaction', 'Age', 'DistanceFromHome', 'NumCompaniesWorked', 'HourlyRate', 'PerformanceRating']


# Create X_df using your selected columns
x_df = attrition_df[columns]

# Show the data types for X_df
x_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.3, random_state=42)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary

# all selected columns is already numeric

x_train.head()
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(x_train)

# Scale the training and testing data
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# Create a OneHotEncoder for the Department column
y_train_dept = y_train['Department']
y_test_dept = y_test['Department']

encoder_dept = OneHotEncoder(sparse=False)

# Fit the encoder to the training data
encoder_dept.fit(y_train_dept.values.reshape(-1,1))

# Create two new variables by applying the encoder
# to the training and testing data
y_train_dept_encoded = encoder_dept.transform(y_train_dept.values.reshape(-1,1))
y_test_dept_encoded = encoder_dept.transform(y_test_dept.values.reshape(-1,1))
```

```python
# Create a OneHotEncoder for the Attrition column
y_train_attr = y_train['Attrition']
y_test_attr = y_test['Attrition']

encoder_attr = OneHotEncoder(sparse=False)

# Fit the encoder to the training data
encoder_attr.fit(y_train_attr.values.reshape(-1,1))

# Create two new variables by applying the encoder
# to the training and testing data
y_train_attr_encoded = encoder_attr.transform(y_train_attr.values.reshape(-1,1))
y_test_attr_encoded = encoder_attr.transform(y_test_attr.values.reshape(-1,1))

```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
n_columns = x_train.shape[1]
print('number of columns ', n_columns)

# Create the input layer
input_layer = layers.Input(shape=n_columns)

# Create at least two shared layers
shared_layer1 = layers.Dense(64, activation='relu', name='shared_layer1')(input_layer)
shared_layer2 = layers.Dense(32, activation='relu', name='shared_layer2')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
hidden_dept = layers.Dense(32, activation='relu', name='hidden_dept')(shared_layer2)

# Create the output layer
output_dept = layers.Dense(len(y_train['Department'].unique()), activation='softmax', name='dept_output')(hidden_dept)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
hidden_attr = layers.Dense(32, activation='relu', name='hidden_attr')(shared_layer2)

# Create the output layer
output_attr = layers.Dense(len(y_train['Attrition'].unique()), activation='sigmoid', name='attr_output')(hidden_attr)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[output_dept, output_attr])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics={'dept_output': 'accuracy', 'attr_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(x_train_scaled, [y_train_dept_encoded, y_train_attr_encoded], epochs=100, batch_size=32)
```

```python
# Evaluate the model with the testing data
results = model.evaluate(x_test_scaled, [y_test_dept_encoded, y_test_attr_encoded])

results
```

```python
# Print the accuracy for both department and attrition
print(f'Department Predictions Accuracy: {results[3]}')
print(f'Attrition Predictions Accuracy: {results[4]}')
```
